```python
import pandas as pd
import requests as req
from bs4 import BeautifulSoup as BS
```

```python
## Create the container of all Products
products_container = []
## create Script for get data from the link
def main(link): 
    page = req.get(link)
    soup = BS(page.text, 'html.parser')
    products = soup.select('.c-prd')
    for prd in products:
        ## get some data from atrribute of the element a with class core
        data_fm_a = prd.select_one('a.core')
        category = data_fm_a.get('data-ga4-item_category2')
        brand = data_fm_a.get('data-gtm-brand')
        prd_name = data_fm_a.get('data-gtm-name')
        ## get data from info div
        info_data = prd.select_one('div.info')
        desc = info_data.select_one('h3.name').text.strip()
        new_price =info_data.select_one('div.prc').text.strip()
        old_data =info_data.select_one('div.s-prc-w')
        if old_data :        
            old_price = old_data.select_one('div.old').text.strip() 
            discount = old_data.select_one('div._dsct' ).text.strip() 
        else: 
            old_price = None
            discount = None

        ### put all data in list of dictionary to use at printing
        products_container.append({
            'Category': category,
            'Brand': brand,
            'Name': prd_name,
            'New_Price': new_price,
            'Discount': discount,
            'Old_Price': old_price,
            
        })
    

    
```

```python
### printin data in file 
def printing():
    for i in range(1,20):
        print(i)
        link = f'https://www.jumia.com.eg/catalog/?q=mobile+phone&page={i}'
        main(link)
printing()
df = pd.DataFrame(products_container)
df.isnull().sum()
```
